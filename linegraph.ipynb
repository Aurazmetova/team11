```python
import pandas as pd
import matplotlib as mlp
import matplotlib.pyplot as plt
usage=pd.read_csv("usage_Uzbekistan.csv")
plt.figure(figsize=(15,8))
plt.title("Usage of different apps in Uzbekistan between 2020-2021",fontdict={"fontweight":"bold","fontsize":18})
# plt.plot(usage.Date,usage.Pinterest,"b.-",label="Pinterest")
# plt.plot(usage.Date,usage.Twitter,"y.-",label="Twitter")
# plt.plot(usage.Date,usage.Facebook,"g.-",label="Facebook")

for apps in usage:
      if apps != "Date":
        plt.plot(usage.Date, usage[apps],marker=".")
    
plt.legend(["Pinterest","Twitter","Facebook","YouTube","Instagram", "VKontakte","TumbIr","reddit","LinkedIn","Fark","newsycombinator.com","Other"])
plt.show()
```

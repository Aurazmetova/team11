```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['16-24', '25-34', '35-44', '45-54', '55-64']
online_via_mobile = [4.10, 3.45, 3.5, 2.22, 3.42]
online_via_PC = [3.30, 3.37, 3.21, 3.23, 3.11]

x = np.arange(len(labels))  
width = 0.35

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, online_via_mobile, width,color="g", label='Online via moble')
rects2 = ax.bar(x + width/2, online_via_PC, width,color="b", label='Online via PC\Laptop')



ax.set_ylabel('Hours')
ax.set_xlabel("Category")
ax.set_title('Avarage daily internet usage worldwide 2019,by age and device',fontdict={"fontweight":"bold","fontsize":14})
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()




fig.tight_layout()

plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

values=[566.14,2525.03,727.85,184.86,332.91,28.92]
regions=["Africa", "Asia", "Europe","Middle East","North America/Caribbean","Oceania/Australia"]
fig = plt.figure(figsize =(10, 7))
colors = ("orange", "cyan", "blue", "green", "indigo", "yellow")
plt.pie(values, labels = regions,colors=colors, autopct="%0.2f%%",shadow=True,explode=[0,0,0,0,0.5,0])
font = {'family': 'serif','color':  'darkred', 'weight': 'normal', 'size': 16,}

plt.title("The percentage of internet users by regions(2020)",fontdict=font)
plt.show()           
```
